## Part I: Working with Dictionary

- **A data structure used to store data in key-value pairs.**
```python
student = {
    "name": "Ram",
    "age": 20,
    "faculty": "BCE"
}
```

### Tasks: Perform following operations

| **Operation** | **Syntax** | **Example** |
|---------------|------------|-------------|
| Create dictionary | `d = {}` | `d = {"a": 1, "b": 2}` |
| Access value | `d[key]` | `d["a"]` |
| Use get() | `d.get(key)` | `d.get("a")` |
| Add item | `d[key] = value` | `d["c"] = 3` |
| Update value | `d[key] = new_value` | `d["a"] = 10` |
| Remove item using del() | `del d[key]` | `del d["b"]` |
| Remove using pop() | `d.pop(key)` | `d.pop("a")` |
| Remove last item | `d.popitem()` | `d.popitem()` |
| Clear dictionary | `d.clear()` | `d.clear()` |
| Copy dictionary | `d.copy()` | `d.copy()` |
| Get all keys | `d.keys()` | `d.keys()` |
| Get all values | `d.values()` | `d.values()` |
| Get all items | `d.items()` | `d.items()` |
| Check key exists | `key in d` | `"a" in d` |
| Get length | `len(d)` | `len(d)` |
| Loop through keys | `for k in d:` | `for k in d:` |
| Loop through values | `for v in d.values():` | `for v in d.values():` |
| Loop through items | `for k, v in d.items():` | `for k, v in d.items():` |

````python
# Create dictionary
student = {
    "name": "Ram",
    "age": 20,
    "faculty": "BCE"
}

student
````

````python
# Access dictionary values
student["name"]
````

````python
# Use get() method to access values
student.get("age")
````

````python
# Add item
student["grade"] = "A"
student
````

````python
# Update value
student["age"] = 21
student
````

````python
# Remove item using del() method
del student["faculty"]
````

````python
# Remove item using pop() method
student.pop("grade")
student
````

````python
# Remove last item using popitem() method
student.popitem()
student
````

````python
# Clear dictionary
student.clear()
student
````

````python
# Add some items again
student["name"] = "Ram"
student["age"] = 21
student
````

````python
# Copy dictionary
new_student = student.copy()
new_student
````

````python
# Get all keys
student.keys()
````

````python
# Get all values
student.values()
````

````python
# Get all items
student.items()
````

````python
# Check if kdy exists
"name" in student
````

````python
#  Get length of dictionary
len(student)
````

````python
# Loop through keys
for key in student:
    print(key)
````

````python
# Loop through values
for value in student.values():
    print(value)
````

````python
# Loop through items
for key, value in student.items():
    print(f"{key}: {value}")
````

## Part II: Working with Pandas DataFrame

A DataFrame in Python is a two-dimensional, tabular data structure that organizes
data into labeled rows and columns. It is conceptually identical to an Excel
spreadsheet or a SQL table, making it the primary tool for data manipulation and
analysis in the Pandas Library.

| Task | Syntax |
|---|---|
| Create a DataFrame containing student details such as Name, Age, and Marks. | `df = pd.DataFrame()` |
| Display the first 5 rows of the DataFrame. | `df.head()` |
| Display the last 5 rows of the DataFrame. | `df.tail()` |
| Find the total number of rows and columns in the DataFrame. | `df.shape` |
| Display all column names in the DataFrame. | `df.columns` |
| Display complete information about the DataFrame including data types and memory usage. | `df.info()` |
| Generate a statistical summary of numerical columns in the DataFrame. | `df.describe()` |
| Select and display the Name column from the DataFrame. | `df['Name']` |
| Select and display multiple columns such as Name and Marks. | `df[['Name', 'Marks']]` |
| Display the first row of the DataFrame using positional indexing. | `df.iloc[0]` |
| Display the first row of the DataFrame using label-based indexing. | `df.loc[0]` |
| Access a specific cell value such as Marks of the first student. | `df.loc[0, 'Marks']` |
| Filter and display students whose Marks are greater than 80. | `df[df['Marks'] > 80]` |
| Filter and display students whose Age is greater than 20 and Marks are greater than 75. | `df[(df['Age'] > 20) & (df['Marks'] > 75)]` |
| Add a new column called Grade to the DataFrame. | `df['Grade'] = ['A', 'B']` |

| Remove the first row from the DataFrame. | `df.drop(0, axis=0)` |

| Sort the DataFrame by Marks in descending order. | `df.sort_values('Marks', ascending=False)` |
| Concatenate two DataFrames vertically. | `pd.concat([df1, df2])` |
| Read data from a CSV file into a DataFrame. | `pd.read_csv('file.csv')` |
| Write the DataFrame into a CSV file. | `df.to_csv('file.csv')` |
| Read data from an Excel file into a DataFrame. | `pd.read_excel('file.xlsx')` |
| Write the DataFrame into an Excel file. | `df.to_excel('file.xlsx')` |

````python
import pandas as pd
````

````python
# Create a dictionary with student details
student_dict = {
    "Name": ["Ram", "Shyam", "Sita", "Gita", "Hari", "Maya", "Rita", "Suresh", "Ramesh", "Sanjay"],
    "Age": [20, 21, 19, 22, 20, 21, 19, 23, 20, 22],
    "Marks": [85, 90, 95, 88, 92, 87, 93, 89, 91, 86]
}

student_dict
````

````python
# Create a DataFrame containing student details such as Name, Age, and Marks from the above dictionary.
dataframe = pd.DataFrame(student_dict)

dataframe
````

````python
# Display the first 5 rows of the DataFrame.
dataframe.head(5)
````

````python
# Display the last 5 rows of the DataFrame.
dataframe.tail(5)
````

````python
#  Find the total number of rows and columns in the DataFrame.
dataframe.shape
````

````python
#  Display all column names in the DataFrame.
dataframe.columns
````

````python
#  Display complete information about the DataFrame including data types and memory usage.
dataframe.info()
````

````python
# Generate a statistical summary of numerical columns in the DataFrame.
dataframe.describe()
````

````python
# Select and display the Name column from the DataFrame.
dataframe["Name"]
````

````python
# Select and display multiple columns such as Name and Marks.
dataframe[["Name", "Marks"]]
````

````python
# Display the first row of the DataFrame using positional indexing. 
dataframe.iloc[0]
````

````python
# Display the first row of the DataFrame using label-based indexing.
dataframe.loc[0]
````

````python
#  Access a specific cell value such as Marks of the first student.
dataframe.loc[0, "Marks"]
````

````python
#  Filter and display students whose Marks are greater than 80
dataframe[dataframe["Marks"] > 80]
````

````python
# Filter and display students whose Age is greater than 20 and Marks are greater than 75
dataframe[(dataframe.Age > 20) & (dataframe.Marks > 75)]
````

````python
# Add a new column called Grade to the DataFrame.
dataframe["Grade"] = ["A", "B", "A", "B", "A", "B", "A", "B", "A", "B"]

dataframe
````

````python
# Increase all students’ Marks by 5.
dataframe["Marks"] = dataframe["Marks"] + 5

dataframe
````

````python
#  Update the Marks of Ram to 95 using a condition.
dataframe.loc[dataframe["Name"] == "Ram", "Marks"] = 95

dataframe
````

````python
# Remove the Age column from the DataFrame
dataframe.drop("Age", axis=1, inplace=True)

dataframe
````

````python
# Remove the first row from the DataFrame.
dataframe.drop(0, axis=0, inplace=True)

dataframe
````

````python
# Sort the DataFrame by Marks in ascending order. 
dataframe.sort_values("Marks", inplace=True)

dataframe
````

````python
# Sort the DataFrame by Marks in descending order.
dataframe.sort_values(by="Marks", ascending=False, inplace=True)

dataframe
````

````python
#  Concatenate two DataFrames vertically.
dataframe2 = pd.DataFrame({
    "Name": ["Anita", "Suresh"],
    "Marks": [88, 92],
    "Grade": ["B", "A"]
})

pd.concat([dataframe, dataframe2])
dataframe
````

````python
# Write the DataFrame into a CSV file.
dataframe.to_csv("./assets/lab02-output.csv", index=False)
````

````python
# Read data from a CSV file into a DataFrame.
df = pd.read_csv("./assets/lab02-output.csv")

df
````

````python
# Write the DataFrame into an Excel file.
import openpyxl

df.to_excel("./assets/lab02-output.xlsx", index=False)
````

````python
# Read data from an Excel file into a DataFrame.
excel_df = pd.read_excel("./assets/lab02-output.xlsx")

excel_df
````
